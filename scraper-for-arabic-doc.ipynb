```python
def getSpecialisationName(pdf,page):
    specialisationFound = False
    descriptionComplete = False

    import re 
    pattern_order = r'[0-9]'
    ar = [""] * 8

    keywords = ["‫اﻹدارة اﻟﻌﺎﻣﺔ ﻟﻠﻤﻨﺎﻫﺞ‬",  "‫ﻫﺪف اﻟﺒﺮﻧﺎﻣﺞ‬", "‫ﻓﺘﺮة اﻟﺘﺪرﻳﺐ‬", "‫اﻟﻤﻬﺎرات اﻟﺘﺨﺼﺼﻴﺔ‬", "‫اﻟﻤﻬﺎرات اﻟﻌﺎﻣﺔ‬", "‫ﺷﻬﺎدات اﺣﺘﺮاﻓﻴﺔ ﻳﻤﻜﻦ ﻣﻌﺎدﻟﺘﻬﺎ‬","‫ﻣﺠﺎﻻت اﻟﻌﻤﻞ‬", "‫اﻹدارة اﻟﻌﺎﻣﺔ ﻟﻠﻤﻨﺎﻫﺞ‬"]
    #in english is general management studies, objective, training period, specialised skills, general skills, certifications, areas of work, general studies(removes the footer from being processed)
    
    keyGenNums = ['04', '03', '02', '01']
    documentSection= 0
    index = 0
    page = page -1
    with open('vocation.txt', 'w') as f:
        f.write("".join(pdf[int(page)]))
    
    generalSkills = [""] * 4
    gindex=0

    training=[""] *3
    tindex=0

    # Using readlines()
    file1 = open('vocation.txt', 'r')
    Lines = file1.readlines()
    front = ""
    back = ""
    counts = 1
    # Strips the newline character
    for line in Lines:
        
        if line != '\n':
            
            if keywords[1] in line:
                documentSection+=1
                index+=1
            elif keywords[2] in line:
                documentSection+=1
                index+=1
            elif keywords[3] in line:
                documentSection+=1
                index+=1
            elif keywords[4] in line:
                documentSection+=1
                index+=1
            elif keywords[5] in line:
                documentSection+=1
                index+=1
            elif keywords[6] in line:
                documentSection+=1
                index+=1
            elif documentSection == 0:
                if not specialisationFound and keywords[0] not in line:
                    ar[index] = line
                    index+=1
                    specialisationFound = True
                elif keywords[0] not in line: 
                    ar[index] += line
            elif keywords[7] in line:
                break
            elif documentSection == 1:
                ar[index] += line
            elif documentSection == 2:
                ar[index] += line
                training[tindex] = line
                tindex+=1
            elif documentSection == 3:
                edited = re.sub(pattern_order, '', line)
                
                if(counts == 1 or counts ==4 or counts==7):
                    front += edited
                elif(counts == 2 or counts ==5 or counts==8):
                    back += edited
                elif(counts == 9):
                    ar[index] += front
                    ar[index] += back
                counts+=1

                
            elif documentSection == 4:
                if not any([x in line for x in keyGenNums]):
                    generalSkills[gindex] = line
                    ar[index] += line
                    gindex+=1
            elif documentSection == 5:
                ar[index] += line
            elif documentSection == 6:
                ar[index] += line
                
    return ar
```

```python
def redactedSpecialisation(pdf,page):
    specialisationFound = False
    descriptionComplete = False

    import re 
    pattern_order = r'[0-9]'
    ar = [""] * 7

    # this pdf is sectioned into parts and is not using a typical encoder, thus the attempt is to use the common sections it is divided into to try and correctly organise the information that can than be organised into a csv

    keywords = ["‫اﻹدارة اﻟﻌﺎﻣﺔ ﻟﻠﻤﻨﺎﻫﺞ‬",  "‫ﻫﺪف اﻟﺒﺮﻧﺎﻣﺞ‬", "‫ﻓﺘﺮة اﻟﺘﺪرﻳﺐ‬", "‫اﻟﻤﻬﺎرات اﻟﺘﺨﺼﺼﻴﺔ‬", "‫اﻟﻤﻬﺎرات اﻟﻌﺎﻣﺔ‬", "‫ﺷﻬﺎدات اﺣﺘﺮاﻓﻴﺔ ﻳﻤﻜﻦ ﻣﻌﺎدﻟﺘﻬﺎ‬","‫ﻣﺠﺎﻻت اﻟﻌﻤﻞ‬", "‫ﻟﻠﺘﻮاﺻﻞ‬"]
    #in english is general management studies, objective, training period, specialised skills, general skills, certifications, areas of work, general studies(removes the footer from being processed)
    #0-7
    keyGenNums = ['04', '03', '02', '01']
    documentSection= 0
    index = 0
    page = page -1
    with open('vocation.txt', 'w') as f:
        f.write("".join(pdf[int(page)]))
    
    generalSkills = [""] * 4
    gindex=0

    training=[""] *3
    tindex=0

    # Using readlines()
    file1 = open('vocation.txt', 'r')
    Lines = file1.readlines()
    front = ""
    back = ""
    counts = 1
    hasPeriod = False

    # Strips the newline character
    for line in Lines:
        
        if line != '\n':
            # bruteforce sectioning into different parts per studies
            if keywords[1] in line:
                documentSection=1
                index+=1
            elif keywords[2] in line:
                documentSection=2
                index+=1
            elif keywords[3] in line:
                documentSection=3
                index+=1
            elif keywords[4] in line:
                documentSection=4
                index+=1
            elif keywords[5] in line:
                documentSection=5
            elif keywords[6] in line:
                index+=1
                documentSection=6

            elif documentSection == 0:
                if not specialisationFound and keywords[0] not in line:
                    ar[index] = line
                    index+=1
                    specialisationFound = True
                elif keywords[0] not in line: 
                    ar[index] += line

            elif keywords[7] in line:
                break
            elif documentSection == 4:
                if not any([x in line for x in keyGenNums]):
                    ar[index] += line
            elif documentSection != 5 and documentSection != 7:
                ar[index] += line
            
    return ar
```
