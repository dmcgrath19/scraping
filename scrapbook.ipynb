```python
def branch_sim(sa_df, eu_df, branch, sa_col, eu_col, sa_title, eu_title):
    eu_filtered = eu_df.loc[eu_df.loc[:,eu_col] == branch]
    sa_filtered = sa_df.loc[sa_df.loc[:,sa_col] == branch]
    eu_titles = eu_filtered.loc[:,eu_title].tolist()
    sa_titles = sa_filtered.loc[:,sa_title].tolist()
    eu_embed = embed.encode(eu_titles)
    sa_embed = embed.encode(sa_titles)
    if len(eu_titles) > 0 and len(sa_titles) > 0:        
        sims = torch.nn.functional.normalize(torch.from_numpy(sa_embed)) @ torch.nn.functional.normalize(torch.from_numpy(eu_embed)).t()
        max_idx = torch.argmax(sims, axis  = 1).numpy()
        prob = pd.DataFrame(np.amax(sims.numpy(), axis = 1), columns = ['prob'])
    else:
        max_idx = np.array([])
        prob = pd.DataFrame([], columns = ['prob'])
    tmp_df = pd.concat([sa_filtered.reset_index(drop=True), eu_filtered.iloc[max_idx].reset_index(drop=True)], axis = 1)
    return pd.concat([tmp_df, prob], axis = 1)
#l3 = branch_sim(map_data, sa_data, '1', 'isco_1', 'major', 'latin_title', 'isco_3_label')
```

```python
# Importing required modules
import PyPDF2

# Creating a pdf file object
pdfFileObj = open('education.pdf','rb')

# Creating a pdf reader object
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

# Getting number of pages in pdf file
pages = pdfReader.numPages

# Loop for reading all the Pages
for i in range(pages):

        # Creating a page object
        pageObj = pdfReader.getPage(i)

        # Printing Page Number
        print("Page No: ",i)

        # Extracting text from page
        # And splitting it into chunks of lines
        text = pageObj.extractText().split("  ")

        # Finally the lines are stored into list
        # For iterating over list a loop is used
        for i in range(len(text)):

                # Printing the line
                # Lines are seprated using "\n"
                if(i == 0):
                     print(text[i],end="\n\n")
                

        # For Seprating the Pages
        print()

# closing the pdf file object
pdfFileObj.close()
```
