**Define a function that generates similarities between isco labels and Saudi labels**

```python
def branch_sim(sa_df, eu_df, branch, sa_col, eu_col, sa_title, eu_title):
    eu_filtered = eu_df.loc[eu_df.loc[:,eu_col] == branch]
    sa_filtered = sa_df.loc[sa_df.loc[:,sa_col] == branch]
    eu_titles = eu_filtered.loc[:,eu_title].tolist()
    sa_titles = sa_filtered.loc[:,sa_title].tolist()
    eu_embed = embed.encode(eu_titles)
    sa_embed = embed.encode(sa_titles)
    if len(eu_titles) > 0 and len(sa_titles) > 0:        
        sims = torch.nn.functional.normalize(torch.from_numpy(sa_embed)) @ torch.nn.functional.normalize(torch.from_numpy(eu_embed)).t()
        max_idx = torch.argmax(sims, axis  = 1).numpy()
        prob = pd.DataFrame(np.amax(sims.numpy(), axis = 1), columns = ['prob'])
    else:
        max_idx = np.array([])
        prob = pd.DataFrame([], columns = ['prob'])
    tmp_df = pd.concat([sa_filtered.reset_index(drop=True), eu_filtered.iloc[max_idx].reset_index(drop=True)], axis = 1)
    return pd.concat([tmp_df, prob], axis = 1)
```

The given Python code defines two functions, `assign_level` and `assign_value`. The `assign_level` function checks if the 'iscoGroup' attribute of the input 'r' is not a string; if true, it returns the 'isco_4' attribute of 'r', otherwise, it returns the 'level1' attribute. The `assign_value` function follows a similar logic, returning the 'isco_4_label' attribute if 'iscoGroup' is not a string, and the 'preferredLabel' attribute otherwise.

```python
def assign_value(r):
    if not isinstance(r.iscoGroup,str):
        return r.isco_4_label
    return r.preferredLabel
```

```python
def assign_level(r):
    if not isinstance(r.iscoGroup,str):
        return r.isco_4
    return r.level1
```

```python
def find_sim(df, col):
    sa_titles = df.loc[:,'latin_title'].values
    eu_titles = df.loc[:,col].values
    eu_embed = embed.encode(eu_titles)
    sa_embed = embed.encode(sa_titles)
    sims = torch.nn.functional.normalize(torch.from_numpy(sa_embed)) @ torch.nn.functional.normalize(torch.from_numpy(eu_embed)).t()
    probs = np.amax(sims.numpy(), axis = 1)
    return probs
```
